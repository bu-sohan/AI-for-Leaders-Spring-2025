# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample dataset
data = {
    'Year': list(range(2010, 2024)),
    'Renewable Share (%)': [12, 14, 18, 22, 25, 28, 30, 33, 37, 40, 44, 46, 50, 53]
}
df = pd.DataFrame(data)


plt.figure(figsize=(8, 5))

# Plot the line
sns.lineplot(x='Year', y='Renewable Share (%)', data=df, color='gray', linewidth=2)

# Highlight the most recent data point
plt.plot(df['Year'].iloc[-1], df['Renewable Share (%)'].iloc[-1], marker='o', color='blue', markersize=8)

# Annotation for emphasis
plt.text(df['Year'].iloc[-1] - 1, df['Renewable Share (%)'].iloc[-1] + 2, 
         f"{df['Renewable Share (%)'].iloc[-1]}%", fontsize=12, color='blue', fontweight='bold')

# Labels and title
plt.xlabel("")
plt.ylabel("Renewable Share (%)", fontsize=12)
plt.title("Growth in Renewable Energy Share Over Time", fontsize=14, fontweight='bold')

# Remove unnecessary chart elements (decluttering)
plt.grid(False)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

# Show the plot
plt.show()
```
